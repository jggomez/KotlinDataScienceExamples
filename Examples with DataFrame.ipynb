```kotlin
val dfFillNull = df
    .fillNulls { Size and Weight and Sweetness and Crunchiness and Juiciness and Ripeness }.with { 0.0 }
    .fillNulls{ Quality }.with { "" }
dfFillNull.filter { Acidity == "Created_by_Nidula_Elgiriyewithana" }
```

```kotlin
dfFillNull.filter { Quality == "good" && Size > 0.1 && Size < 0.2 }
```

```kotlin
dfFillNull.add("NewColumn") { Size + Weight }
```

```kotlin
dfFillNull
    .groupBy { Quality }.count()
```

```kotlin
dfFillNull
    .groupBy { Quality }
    .aggregate {
        count { Size > 0 && Size < 1.0 } into "Best Size"
        count { Size > 1 && Size < 2.0 } into "Regular Size"
    }
```
