```kotlin
val df = DataFrame.read("https://storage.googleapis.com/screenshot-devhack-3f0c2/apple_quality.csv")
```

```kotlin
val dfFillNull = df
    .fillNulls { Size and Weight and Sweetness and Crunchiness and Juiciness and Ripeness }.with { 0.0 }
    .fillNulls{ Quality }.with { "" }
dfFillNull.filter { Acidity == "Created_by_Nidula_Elgiriyewithana" }
```

```kotlin
dfFillNull.plot {
    layout.title = "Calificación de Manzanas segun el Tamaño"
    bars {
        x(Quality) { axis.name = "Calidad" }
        y(Size)
        fillColor = Color.ORANGE
    }
}
```

```kotlin
dfFillNull.plot {
    statBin("Size") {
        bars {
            x(Stat.x)
            y(Stat.count)
        }
        line {
            x(Stat.x)
            y(Stat.count)
            color = Color.RED
            width = 1.5
        }
    }
}
```

```kotlin
dfFillNull.groupBy { Quality }.plot {
    layout.title = "Calificación de Manzanas segun el Tamaño"
    histogram("Size")
}
```

```kotlin
dfFillNull.plot { 
    countPlot("Quality")
}
```

```kotlin
dfFillNull.plot {
    points{
        x("Size")
        y("Weight")
    }
}
```

```kotlin
val random = java.util.Random(42)

val sampleA = List(1000) { random.nextGaussian() * 0.7 + 2.0 }
val sampleB = List(1000) { random.nextGaussian() * 1.4 + 3.5 }

val dfSample = dataFrameOf(
    "sample" to sampleA + sampleB,
    "group" to sampleA.map { "A" } + sampleB.map { "B" }
)

dfSample.groupBy("group").plot {
    histogram("sample")
}
```
